### Data Wrangling

```python
# Rename the columns of the DataFrame
df.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Region']
# Convert the 'Frequency' column to a categorical data type
df['Frequency'] = df['Frequency'].astype('category')

# Convert the 'Region' column to a categorical data type
df['Region'] = df['Region'].astype('category')
# Convert the 'Date' column to datetime format, specifying 'dayfirst' to handle date formats with day first
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Extract the month from the 'Date' column and create a new 'Month' column
df['Month'] = df['Date'].dt.month

# Create a new 'Month_int' column by converting the 'Month' column to integers, handling missing values
df['Month_int'] = df['Month'].apply(lambda x: int(x) if pd.notna(x) else x)
# Drop the 'Month' column from the DataFrame
df.drop(columns='Month', inplace=True)
```

```python
# Lets create a copy of the dataset for the data of during lockdown period
df_ld = df.copy()
# Drop all the missing values from 'df' dataset
df=df.dropna()
# Define a date range mask to filter rows where 'Date' falls within the range of before lockdown period
mask = (df['Date'] >= '2019-05-31') & (df['Date'] <= '2020-03-31')

# Use the mask to select and keep only the rows that meet the date range criteria
df = df.loc[mask]

# Define a date range mask to filter rows where 'Date' falls within the range of during lockdown period
mask = (df_ld['Date'] >= '2020-04-30')

# Use the mask to select and keep only the rows that meet the date range criteria
df_ld = df_ld.loc[mask]
```

#### Rename the DataFrame columns and convert **Frequency** and **Region** to categorical data types. Transform the **Date** column to datetime format, extract the month, and create a new **Month_int** column as integers before dropping the original **Month** column. A separate dataset is created for the **during lockdown** period, while missing values are removed from the **before lockdown** dataset. The **before lockdown** period is defined as **2019-05-31 to 2020-03-31**, and the **during lockdown** period includes dates after **2020-04-30**.

### 1. Region Wise Estimated Unemployment Rate

```python
# Chart - 1: Bar plot visualization for Region-wise Estimated Unemployment Rate before and during lockdown

import matplotlib.pyplot as plt
import seaborn as sns

# Create a figure with two subplots side by side.
plt.figure(figsize=(14, 5))

# Subplot 1: Estimated Unemployment Rate Before Lockdown
plt.subplot(1, 2, 1)

# Bar plot with red bars
sns.barplot(x='Region', y='Estimated Unemployment Rate', data=df, palette="Reds_r")

# Set labels and title for the first subplot.
plt.xlabel("Region", fontsize=10)
plt.ylabel('Estimated Unemployment Rate', fontsize=10)
plt.title('Estimated Unemployment Rate Before Lockdown', fontsize=12)

# Subplot 2: Estimated Unemployment Rate During Lockdown
plt.subplot(1, 2, 2)

# Bar plot with blue bars
sns.barplot(x='Region', y='Estimated Unemployment Rate', data=df_ld, palette="Blues_r")

# Set labels and title for the second subplot.
plt.xlabel("Region", fontsize=10)
plt.ylabel('Estimated Unemployment Rate', fontsize=10)
plt.title('Estimated Unemployment Rate During Lockdown', fontsize=12)

# Display the subplots side by side.
plt.show()
```

#### *This visualization presents a comparative analysis of the Estimated Unemployment Rate across different regions before and during the lockdown period using bar plots.*
#### - The left bar chart (red-colored bars) represents the unemployment rate before the lockdown, providing insights into regional employment conditions under normal circumstances.

#### - The right bar chart (blue-colored bars) displays the unemployment rate during the lockdown, highlighting the impact of restrictions on employment across different regions.

#### **By visually comparing both charts, we can observe the change in unemployment rates due to the lockdown and identify which regions were most affected. The use of red and blue colors effectively distinguishes the two time periods, making it easier to analyze the differences in employment trends.**

#### **During Lockdown:**

#### - The second plot (blue color scheme) shows unemployment rates during the lockdown.

#### - A noticeable shift can be observed, with several states experiencing a surge in unemployment rates.

###  5 : Correlation Heatmap

```python
# Create new DataFrames
df1_stats = df[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']]

df2_stats = df_ld[['Estimated Unemployment Rate','Estimated Employed', 'Estimated Labour Participation Rate']]

# Chart - 5 Correlation heatmap visualization code

# Create a figure with two subplots side by side.
plt.figure(figsize=(25, 8))

# Subplot 1: Correlations of Variables Before Lockdown.
plt.subplot(1, 2, 1)

# Set the title for the first subplot.
plt.title("Correlations of Variables Before Lockdown", fontsize=16)

# Create a heatmap to visualize the correlations between variables in df1_stats.
sns.heatmap(df1_stats.corr(), annot=True,cmap="coolwarm", linewidths=0.1, fmt='.2f', square=True)

# Subplot 2: Correlations of Variables During Lockdown.
plt.subplot(1, 2, 2)

# Set the title for the second subplot.
plt.title("Correlations of Variables During Lockdown", fontsize=16)

# Create a heatmap to visualize the correlations between variables in df2_stats.
sns.heatmap(df2_stats.corr(), annot=True,cmap="coolwarm", linewidths=0.1, fmt='.2f', square=True)

# Display the subplots side by side.
plt.show()  
```

### This dual heatmap visualization illustrates the correlation between key employment metrics—Estimated Unemployment Rate, Estimated Employed, and Estimated Labour Participation Rate—before and during the lockdown. The intensity of colors represents the strength and direction of relationships between these variables.

#### **Before Lockdown (Left Heatmap):**

#### Shows how employment, unemployment, and labour participation were interrelated under normal economic conditions.

#### **Expected correlations:**

#### - Negative correlation between Unemployment Rate and Estimated Employed (higher employment, lower unemployment).

#### - Positive correlation between Labour Participation Rate and Estimated Employed (higher participation, more employment opportunities).

### **During Lockdown (Right Heatmap):**

#### Captures how economic disruptions altered these relationships.

### **Potential shifts:**

#### - Weaker employment-labour participation correlation, as job availability declined despite participation.

#### - Stronger unemployment-employment inverse correlation, highlighting job losses.

### **Key Insights:**

#### - Did Unemployment Spike? Higher negative correlation suggests widespread job losses.
#### - Did Participation Drop? If correlation weakened, it may indicate people leaving the workforce.
#### - Economic Resilience? A stable correlation pattern suggests states that maintained labour market stability.

### This analysis helps policymakers and economists understand how the employment landscape evolved during the crisis and what structural changes occurred in the job market.
